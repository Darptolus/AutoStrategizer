```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

n_exp = 6
n_iter = 12
n_sizes = 8
i_sizes = 0
i_exp = 0
t_exp = []
labels = ['(a) Host-to-all (seq)', '(b) Host-to-all (par)', '(c) Main-to-local-v1', '(d) Main-to-local-v2', '(e) Main-to-local-v3', '(f) Splited-linked-list']
xlabels = ['0', '1 Kb',	'10 Kb',	'100 Kb',	'1 Mb',	'10 Mb',	'100 Mb',	'1 Gb',	'5 Gb']

# Initialize lists
for i in range(n_exp):
  sublist = []
  t_exp.append(sublist)

t_plota = [[0 for i in range(n_sizes)] for j in range(n_exp)]
t_plotb = [[0 for i in range(n_sizes)] for j in range(n_exp)]

# Get file
filename = "./results/broadcast_2023_09_29-233437.txt"
data = pd.read_csv(filename, sep='\t')

# Get information
for i_sizes in range(n_sizes):
  i_exp = 0
  for j in range(n_iter):
    for sublist in t_exp:
      sublist.append(float(data.loc[i_exp].iat[i_sizes]))
      i_exp += 1
    i_exp += 1

# Remove 2 outliers and average
  i_exp = 0
  for sublist in t_exp:
    sublist.remove(max(sublist))
    sublist.remove(max(sublist))
    lst_avg = round(numpy.average(sublist), 7)
    t_plota[i_exp][i_sizes] = lst_avg
    i_exp += 1

# Clear lists
  for sublist in t_exp:
    sublist.clear()

# Calculate speedup
for i_exp in range(n_exp):
  for i_sizes in range(n_sizes):
    t_plotb[i_exp][i_sizes] = t_plota[0][i_sizes]/t_plota[i_exp][i_sizes]
```

```python
# Plot Execution Time
fig, ax = plt.subplots(figsize=(8,4))
ax.grid(axis='y', alpha=0.5)

ax.set_xticklabels(xlabels)

for i, sublist in enumerate(t_plota):
  plt.plot(sublist, label=labels[i], marker=".")

# Add labels and title
plt.xlabel('Payload Size')
plt.ylabel('Execution Time (Sec)')
plt.title('Broadcasting (8 GPUs)')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Plot Speedup
fig, ax = plt.subplots(figsize=(8,4))
ax.grid(axis='y', alpha=0.5)
ax.set_xticklabels(xlabels)

for i, sublist in enumerate(t_plotb):
  plt.plot(sublist, label=labels[i], marker=".")

# Add labels and title
plt.xlabel('Payload Size')
plt.ylabel('Speedup')
plt.title('Broadcast (8 GPUs) - Speedup')

# Add legend
plt.legend()

plt.savefig('Broadcast.png', bbox_inches='tight')
# Display the plot
plt.show()
```
